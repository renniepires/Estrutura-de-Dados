```python
# 1. Faça um programa que calcule a média de três números inseridos pelo usuário.

nota1 = float(input('Insira o valor da primeira nota: '))
nota2 = float(input('Insira o valor da segunda nota: '))
nota3 = float(input('Insira o valor da terceira nota: '))
media = (nota1 + nota2 + nota3) / 3

print(f'As suas notas foram: {nota1}, {nota2} e {nota3}.')
print(f'O resultado da media das notas inseridas é igual a {media:.1f}.')

if media >= 7:
    print('Parabéns, você foi APROVADO!')
else:
    print('Desculpe, você foi REPROVADO.')
```

```python
# 4. Crie um programa que leia uma lista de números e exiba o maior e o menor valor da lista.

def lista():
    try:
        lista_numeros = input("Insira uma lista de números inteiros separados por espaços: ")
        numeros = [int(num) for num in lista_numeros.split()]

        if not numeros:
            print("A lista está vazia.")
            return

        valor_max = max(numeros)
        valor_min = min(numeros)

        print("Maior valor:", valor_max)
        print("Menor valor:", valor_min)

    except ValueError:
        print("Entrada inválida. Certifique-se de digitar números inteiros separados por espaços.")

if __name__ == "__main__":
    lista()
```

```python
# 5. Faça um programa que leia uma lista de números e retorne a média dos números pares.

def media_numeros_pares(lista):
    numeros_pares = [num for num in lista if num % 2 == 0]

    if not numeros_pares:
        return 0 

    soma_pares = sum(numeros_pares)
    media = soma_pares / len(numeros_pares)
    return media

def main():

    try:
        input_numbers = input("Digite uma lista de números inteiros separados por espaços: ")

        numbers = [int(num) for num in input_numbers.split()]

        media_pares = media_numeros_pares(numbers)

        if media_pares == 0:
            print("Não há números pares na lista.")
        else:
            print("Média dos números pares:", media_pares)

    except ValueError:
        print("Entrada inválida. Certifique-se de digitar números inteiros separados por espaços.")

if __name__ == "__main__":
    main()
```

```python
# 6. Escreva um programa que peça um número inteiro positivo ao usuário e calcule o fatorial desse número.

def fatorial(numero):
    if numero == 0 or numero == 1:
        return 1
    else:
        return numero * fatorial(numero - 1)

def main():
    try:
        numero = int(input("Digite um número inteiro positivo: "))
        
        if numero < 0:
            print("O número deve ser positivo.")
        else:
            fatorial = fatorial(numero)
            print(f"O fatorial de {numero} é {fatorial}")

    except ValueError:
        print("Entrada inválida. Certifique-se de digitar um número inteiro.")

if __name__ == "__main__":
    main()
```

```python
# 7. Crie um programa que imprima a sequência de Fibonacci até um valor inserido pelo usuário.

def sequencia_fibonacci(n):
    sequencia = []
    a, b = 0, 1
    while a <= n:
        sequencia.append(a)
        a, b = b, a + b
    return sequencia

def fibonacci():
    try:
        limite = int(input("Digite o valor limite para a sequência de Fibonacci: "))
        
        if limite < 0:
            print("O valor limite deve ser positivo.")
        else:
            seq_fib = sequencia_fibonacci(limite)
            print("Sequência de Fibonacci até", limite, ":", seq_fib)

    except ValueError:
        print("Entrada inválida. Certifique-se de digitar um número inteiro.")

if __name__ == "__main__":
    fibonacci()
```

```python
# 8. Faça um programa que determine se um número é primo ou não.

def numeros_primos(numero):
    if numero <= 1:
        return False
    if numero <= 3:
        return True
    if numero % 2 == 0 or numero % 3 == 0:
        return False
    i = 5
    while i * i <= numero:
        if numero % i == 0 or numero % (i + 2) == 0:
            return False
        i += 6
    return True

def detectar():
    try:
        numero = int(input("Digite um número inteiro: "))
        
        if numeros_primos(numero):
            print(numero, "é um número primo.")
        else:
            print(numero, "não é um número primo.")

    except ValueError:
        print("Entrada inválida. Certifique-se de digitar um número inteiro.")

if __name__ == "__main__":
    detectar()
```

```python
# 9. Escreva um programa que leia uma lista de nomes e retorne uma nova lista com apenas os nomes que começam com a letra 'A'.

def inicial_a():
    try:
        # Solicitar ao usuário a entrada dos nomes separados por espaços
        nomes_input = input("Digite uma lista de nomes separados por espaços: ")

        # Dividir a entrada em uma lista de strings
        nomes = nomes_input.split()

        # Filtrar os nomes que começam com a letra 'A'
        nomes_a = [inicial for inicial in nomes if inicial.startswith('A') or inicial.startswith('a')]

        # Exibir os resultados
        if not nomes_a:
            print("Não foram encontrados nomes que comecem com 'A' na lista.")
        else:
            print("Nomes que começam com 'A':", nomes_a)

    except ValueError:
        print("Entrada inválida.")

if __name__ == "__main__":
    inicial_a()
```

```python
# 10. Crie um programa que simule o jogo "Pedra, Papel e Tesoura" entre o usuário e o computador. O programa deve solicitar a escolha do usuário e, em seguida, escolher aleatoriamente a escolha do computador e determinar o vencedor.

import random

def computar_escolha_do_computador():
    escolhas = ["Pedra", "Papel", "Tesoura"]
    return random.choice(escolhas)

def determine_winner(escolha_usuario, escolha_computador):
    if escolha_usuario == escolha_computador:
        return "Empate"
    elif (escolha_usuario == "Pedra" and escolha_computador == "Tesoura") or \
         (escolha_usuario == "Papel" and escolha_computador == "Pedra") or \
         (escolha_usuario == "Tesoura" and escolha_computador == "Papel"):
        return "Você venceu!"
    else:
        return "Você perdeu!"

def partida():
    try:
        escolha_usuario = input("Escolha Pedra, Papel ou Tesoura: ")
        escolha_usuario = escolha_usuario.capitalize()

        if escolha_usuario not in ["Pedra", "Papel", "Tesoura"]:
            print("Escolha inválida. Escolha entre Pedra, Papel ou Tesoura.")
        else:
            escolha_computador = computar_escolha_do_computador()
            print("Computador escolheu:", escolha_computador)
            resultado = determine_winner(escolha_usuario, escolha_computador)
            print(resultado)

    except ValueError:
        print("Entrada inválida.")

if __name__ == "__main__":
    partida()
```
